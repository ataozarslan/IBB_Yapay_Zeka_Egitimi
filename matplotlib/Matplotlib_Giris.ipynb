## 2. Temel Grafik Türleri

### 2.1. Bar Chart (Çubuk Grafik)

Çubuk grafikleri, Verileri kategoriler arasında karşılaştırmak için kullanılırlar. Çubuk grafikler aynı zamanda çok bilinir bir grafik türü olduğu için, kurallarına uygun olarak oluşturulursa okunması ve yorumlanması oldukça kolaydır. 

Bu çubuk grafik, her bir iş sektöründe alınan kredilerin sayısını ölçer. İyi bir çubuk grafiği oluşturmak için,
- Tek bir renk kullanın (veriniz belli gruplardan oluşmuyorsa)
- Veriler en yüksekten en düşüğe doğru sıralayın
- Etiket kullanıyorsanız, okunabilir olmasına dikkat edin
- Eksenleri sıfırdan başlatın

![bar_chart.png](https://www.tableau.com/sites/default/files/2021-08/Bar%20Chart%20-%20Good%20-%20900x650.png)

```python
new_x = np.arange(20)
new_y = np.random.randint(5,10,20)
plt.bar(new_x, new_y);
```

```python
new_x = np.arange(20)
new_y = np.random.randint(5,10,20)
plt.barh(new_x, new_y);
```

### 2.2. Histogram

Histogramlar, çubuk grafiklerin belirli bir varyasyonudur ve veri dağılımlarını göstermenin bir yolunu sağlar ve bir veri kümesinde bulunan farklı değerleri bir dizi bağlantılı çubuk olarak izler.

Bu histogram, Austin, Teksas'taki Airbnb kiralamalarına bakar ve günlük fiyatı 25 ABD doları cinsinden gösterir. Grafik sağa çarpık bir dağılıma sahip ve bir Airbnb'nin ortalama fiyatı gecelik 50$ ile 150$ arasında görünüyor. Bu histogram yalnızca bir renk kullanır Tek bir ölçüye bakar Kolayca tahmin edilebilen bir ortalamaya sahiptir.

![histogram.png](https://www.tableau.com/sites/default/files/2021-09/Histogram%20-%20Good%20-%20900x650%20%E2%80%93%20UPDATED%20%E2%80%93%202.png)

```python
new_x = np.random.randn(100000)
plt.hist(new_x);
```

```python
new_x = np.random.randn(100000)
plt.hist(new_x, bins=25); # Bins parametresi histogramı kaç parça olarak görmek istediğimizi belirler
```

### 2.3. Pie Chart (Pasta Grafik)

Pasta grafiği, verileri bir bütünün yüzdesi olarak düzenlemeye ve göstermeye yardımcı olur. Adına uygun olarak, bu tür görselleştirme, bütünü temsil etmek için bir daire ve bütünü oluşturan belirli kategorileri temsil etmek için bu dairenin dilimleri veya "pasta" kullanır.

Bu pasta grafik, bölgesel pazar başına toplam satışların yüzdesini gösterir. İyi bir pasta grafiği oluşturmak için,
- Beş kategori diliminden fazla kullanmayın 
- En büyük dilimi dairenin tepesinden başlatın
- Dilimleri belirgin, canlı renklerle renklendirin (mümkünse renk seçimini içeriğe göre yapın)
- Renk körü kullanıcılara uygun renkleri kullanmayı ihmal etmeyin

![pie_charts.png](https://www.tableau.com/sites/default/files/2021-06/Pie%20Chart%201%20-%20Good%20-%20900x650.png)

```python
new_x = np.arange(1,4)
plt.pie(new_x, labels=['a','b','c']);
```

```python
plt.pie(new_x, labels=['a','b','c'], explode = [.1,.2,0]);
```

### 2.4. Line Chart (Çizgi Grafik)

Çizgi grafiği, bir çizgi kullanarak bir dizi veri noktasını birbirine bağlar. Bu grafik türü, eğilimleri belirlemenize yardımcı olacak sıralı değerler sunar. Çoğu zaman, x ekseni (yatay eksen), değerlerin sıralı bir ilerlemesini temsil eder. Y ekseni (dikey eksen) daha sonra size o ilerleme boyunca seçilen bir metrik için değerleri söyler. Bu yaygın bir grafiktir ve zaman içindeki verileri göstermek istediğinizde kullanmak harikadır.

Kategori çizgi grafiği olarak bilinen bu çizgi grafik, borçluların sayısını cinsiyete böler, böylece birkaç yıl boyunca iki çizgi görüntülenir. İyi bir çizgi grafiği oluşturmak için,

- X eksenini tarih alanını temsil etmek için, Y eksenini nicel alanı temsil etmek için kullanın
- Odaklanılan kategorinin çizgisini vurgulamak için güçlü, diğer kategori için de nötr bir renk kullanın
- Etiket renklerinin, kolay bir şekilde tanınabilmesi için lejant eklemeyi unutmayın

![line_charts.png](https://www.tableau.com/sites/default/files/2021-09/Line%20Chart%202%20-%20Good%20-%20900x650.png)

```python
new_x = np.arange(100)
new_y = np.random.randint(0,10,100)
plt.plot(new_x, new_y);
```

### 2.5. Scatter Plot (Dağılım Grafiği)

Dağılım grafiği, iki ölçü veya nicel değişken arasındaki ilişkiyi incelemek istediğimizde kullanışlıdır. Değişkenlerin arasında
bir ilişki var mı, varsa ne ölçüde şeklindeki sorulara cevap verir. Ayrıca veri setinde bulanabilecek olan aykırı değerler hakkında bilgi verebilir.

Bu dağılım grafiği, veri seti içerisinde bulunan kişilerin boy ve kiloları arasındaki ilişkiyi gösterir. İyi bir dağılım grafiği oluşturmak için,

- X ve Y ekseninin başlangıç ve bitiş noktalarına elinizdeki verinin minimum ve maksimum değerlerine göre karar verin
- Eğer gerekliyse noktalar üzerine renk veya şekil gibi eklentiler yaparak 3. bir özellik hakkında da bilgi sağlayın

![scatter_plots.png](https://interworks.com/wp-content/uploads/sites/default/files/TECTScatter3.png)

```python
new_x = np.random.randn(1000)
new_y = np.random.randn(1000)
plt.scatter(new_x, new_y);
```

```python
plt.scatter(new_x, new_y, alpha=0.3);
```

### 2.6. Heatmap (Isı Haritaları)

Isı haritaları, metin tablosu olarak görüntülenen verileri görselleştirmeye yardımcı olmak için renk kullanır. Ölçü değeri aralıklarını, rengin dikkat çekici özelliğiyle kodlarlar; en düşükten en yükseğe. Bu tablolar, sıralı veya farklı paletler kullanarak sürekli renkleri görüntüleyebilir.

Bu öne çıkan ısı haritası, bölgelere göre ve ürün satışlarına ait bilgiler sunar. Renk ne kadar koyu olursa satış miktarı o kadar yüksek, ne kadar açık olursa da satış miktarının o kadar düşüktür. İyi bir ısı haritası oluşturmak için,

- Değişkenler arasındaki büyüklük/küçüklük ilişkisini yansıtabilecek renkler paletleri kullanın

### 2.7. Box Plot (Kutu Grafik)

Kutu grafikleri, seçili bir hesaplama genelinde veri noktalarının dağılımını gösterirken kullanmak için harika bir grafiktir. Bu grafikler, ölçülen değişkenler içindeki aralıkları gösterir. Ek olarak veri setinde bulunan aykırı değerlerin tespitinde en sık kullanılan yöntemlerden biridir.

![quartiles.png](https://help.ezbiocloud.net/wp-content/uploads/2020/04/%E1%84%89%E1%85%B3%E1%84%8F%E1%85%B3%E1%84%85%E1%85%B5%E1%86%AB%E1%84%89%E1%85%A3%E1%86%BA-2020-04-19-%E1%84%8B%E1%85%A9%E1%84%8C%E1%85%A5%E1%86%AB-9.41.35.png)

Bu kutu grafiğinde, yarı zamanlı roller ve tam zamanlı roller için maaşlar analiz edilir. İki grafik, her tür iş için ortalamayı gösterir. Başvuru sahipleri, belirli bir iş türü için ne beklemeleri gerektiğini öğrenebilir ve analistler, hangi iş unvanlarının aykırı olduğunu hızlı bir şekilde belirleyebilir. İyi bir kutu grafiği oluşturmak için,

- Ölçülerde tutarlı renkler kullanın
- %25, %50 (medyan) ve %75'lik dilimleri açıkça belirtin

![box_plots.png](https://www.tableau.com/sites/default/files/2021-09/Box%20_%20Whisker%20-%20Good%20-%20900x650.png)

```python
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
data = np.concatenate((spread, center, flier_high, flier_low))
plt.boxplot(data);
```

### 3.9. Eksen Aralığı

Grafikte oluşturulan eksenlerin başlangıç ve bitiş noktalarını değiştirmek için `plt.axis()` fonksiyonundan yararlanabiliriz.

### 4. Çoklu Grafikler Oluşturma

Bazen çalışmalarımızda yan yana birden çok grafik göstermek isteyebiliriz. Bu konuda da yardımımıza `plt.subplot` fonksiyonu yetişiyor.

```python
new_x = np.arange(20)
new_y = np.random.randint(5,10,20)

plt.figure(figsize=[18,12])

plt.suptitle('Main Title', fontsize = 16)

plt.subplot(2,2,1) # (Satır Sayısı, Sütun Sayısı, Grafik No)
plt.plot(new_x, new_y)
plt.title('Line Chart')

plt.subplot(2,2,2)
plt.bar(new_x, new_y)
plt.title('Bar Chart');

plt.subplot(2,2,3)
plt.bar(new_x, new_y)
plt.title('Bar Chart');

plt.subplot(2,2,4)
plt.barh(new_x, new_y)
plt.title('Bar Chart');
```

```python
new_x = np.arange(20)
new_y = np.random.randint(5,10,20)

plt.figure(figsize=[15,10])

plt.suptitle('Main Title', fontsize = 16)

plt.subplot(2,1,1)
plt.plot(new_x, new_y)
plt.title('Line Chart')

plt.subplot(2,1,2)
plt.bar(new_x, new_y)
plt.title('Bar Chart');
```
